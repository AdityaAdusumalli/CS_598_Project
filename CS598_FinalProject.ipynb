```python

utils_content = r"""
import os
import sys
import ast
import pickle
import gc # For garbage collection

import wfdb
import numpy as np
import pandas as pd
from tqdm import tqdm
from sklearn.preprocessing import StandardScaler, MultiLabelBinarizer

sys.path.append(os.path.join(os.path.dirname(__file__), ".."))
import config

def form_datasplit_string(split_number: int) -> str:
    fold_indices = config.split_settings[split_number]
    val_fold_index = fold_indices["val_index"]
    test_fold_index = fold_indices["test_index"]
    data_split_string = f"val-{val_fold_index}_test-{test_fold_index}"
    return data_split_string

def load_dataset(path, sampling_rate, release=False):
    X = None
    Y = None

    norm_path = os.path.normpath(path)
    dataset_folder_name = os.path.basename(os.path.dirname(norm_path))
    print(f"load_dataset: Normalized path: {norm_path}, Detected dataset folder: {dataset_folder_name}") # DEBUG

    if dataset_folder_name == config.dirname_ptbxl:
        ptbxl_csv_file = os.path.join(path, 'ptbxl_database.csv')
        print(f"load_dataset: Checking for PTB-XL CSV at {ptbxl_csv_file}") # DEBUG
        if os.path.exists(ptbxl_csv_file):
            Y_full = pd.read_csv(ptbxl_csv_file, index_col='ecg_id') # Load full Y
            Y_full.scp_codes = Y_full.scp_codes.apply(lambda x: ast.literal_eval(x))
            print("load_dataset: PTB-XL CSV loaded, attempting to load raw signal data...") # DEBUG

            # This is the hardcoded subset size for debugging
            # It's passed to load_raw_data_ptbxl
            debug_subset_size = 1000

            X = load_raw_data_ptbxl(Y_full, sampling_rate, path, subset_size=debug_subset_size)

            # If X was successfully subsetted, Y must also be subsetted to match
            if X is not None and len(X) == debug_subset_size:
                print(f"load_dataset: X was subsetted to {len(X)} records. Subsetting Y (labels) to match.") #DEBUG
                Y = Y_full.head(debug_subset_size).copy() # Use .copy() to avoid SettingWithCopyWarning
            else:
                # If X is not subsetted (e.g., subset_size was None in load_raw_data_ptbxl or loading failed)
                # then use the full Y. For this debug path, we expect subsetting.
                # If X is None, an error will be raised later anyway.
                print(f"load_dataset: X was NOT subsetted as expected (len(X)={len(X) if X is not None else 'None'}, expected_subset={debug_subset_size}). Using full Y. This might cause downstream errors if X is subsetted and Y is not.") # DEBUG
                Y = Y_full

            print("load_dataset: Raw signal data loading function finished.") # DEBUG
        else:
            print(f"ERROR: PTB-XL database CSV not found at {ptbxl_csv_file}", file=sys.stderr)

    elif dataset_folder_name == config.dirname_g12ec:
        icbeb_csv_file = os.path.join(path, 'icbeb_database.csv')
        if os.path.exists(icbeb_csv_file):
            Y = pd.read_csv(icbeb_csv_file, index_col='ecg_id')
            Y.scp_codes = Y.scp_codes.apply(lambda x: ast.literal_eval(x))
            print(f"WARNING: Matched G12EC dataset and loaded CSV, but 'load_raw_data_icbeb' or equivalent for G12EC raw signals is missing in utils.py. Raw signal data (X) will be None.", file=sys.stderr)
            X = None
        else:
            print(f"ERROR: G12EC/ICBEB database CSV ('icbeb_database.csv') not found at {icbeb_csv_file}", file=sys.stderr)

    elif dataset_folder_name == config.dirname_cpsc:
        print(f"WARNING: Matched CPSC2018 dataset folder, but utils.load_dataset doesn't have specific loading logic for it. X and Y will be None.", file=sys.stderr)
        X = None
        Y = None

    if X is None or Y is None:
        error_message = f"Failed to load dataset. X or Y is None. Path provided: {path} (normalized: {norm_path}).\\n"
        error_message += f"Detected dataset folder name from path: '{dataset_folder_name}'\\n"
        error_message += f"Expected PTB-XL folder: '{config.dirname_ptbxl}', G12EC folder: '{config.dirname_g12ec}', CPSC folder: '{config.dirname_cpsc}'.\\n"
        if dataset_folder_name == config.dirname_ptbxl:
             error_message += f"PTB-XL CSV check: '{os.path.join(path, 'ptbxl_database.csv')}' exists? {os.path.exists(os.path.join(path, 'ptbxl_database.csv'))}\\n"
        elif dataset_folder_name == config.dirname_g12ec:
             error_message += f"G12EC/ICBEB CSV check: '{os.path.join(path, 'icbeb_database.csv')}' exists? {os.path.exists(os.path.join(path, 'icbeb_database.csv'))}\\n"
        error_message += "Please ensure path is correct, necessary data/CSV files exist, and relevant loading functions are defined in utils.py."
        raise ValueError(error_message)

    return X, Y

def load_raw_data_ptbxl(df, sampling_rate, path, subset_size=None): # Added subset_size
    data = None

    df_to_process = df
    if subset_size is not None:
        print(f"load_raw_data_ptbxl: Processing a subset of {subset_size} records for debugging.") # DEBUG
        df_to_process = df.head(subset_size)

    if sampling_rate == 100:
        # Modify npy path if subset_size is used to avoid overwriting full data / using partial data
        npy_filename = f'raw100_subset{subset_size}.npy' if subset_size else 'raw100.npy'
        raw_npy_path = os.path.join(path, npy_filename)

        if os.path.exists(raw_npy_path) and subset_size is None: # Only load if not a subset or if subset file exists
            print(f"load_raw_data_ptbxl: Loading existing {raw_npy_path}") # DEBUG
            data = np.load(raw_npy_path, allow_pickle=True)
        else:
            print(f"load_raw_data_ptbxl: Reading signals for {raw_npy_path} (subset_size: {subset_size})...") # DEBUG
            signals_list = []
            for i, f_path_suffix in enumerate(tqdm(df_to_process.filename_lr, desc=f"Loading 100Hz signals (subset {subset_size})" if subset_size else "Loading 100Hz signals")):
                full_file_path = os.path.join(path, f_path_suffix)
                try:
                    signal_data, meta_data = wfdb.rdsamp(full_file_path)
                    signals_list.append(signal_data)
                except Exception as e:
                    print(f"Error reading file {i} ({full_file_path}): {e}", file=sys.stderr)
                    signals_list.append(None)

            print(f"load_raw_data_ptbxl: Finished reading {len(signals_list)} signals. Filtering None entries...") # DEBUG
            valid_signals = [s for s in signals_list if s is not None]
            del signals_list # Free memory
            gc.collect() # Force garbage collection

            if not valid_signals:
                print("WARNING: No valid signals were loaded for 100Hz. `data` will be an empty array.", file=sys.stderr)
                data = np.array([], dtype=object)
            else:
                print(f"load_raw_data_ptbxl: {len(valid_signals)} valid signals. Attempting conversion to numpy array...") # DEBUG
                try:
                    data = np.array(valid_signals, dtype=object)
                    print(f"load_raw_data_ptbxl: Conversion to numpy array complete. Shape: {data.shape if hasattr(data, 'shape') else 'N/A'}. Pickling...") # DEBUG
                except MemoryError:
                    print("load_raw_data_ptbxl: MEMORY ERROR during np.array conversion for 100Hz. Try reducing subset_size or using a high-RAM runtime.", file=sys.stderr)
                    raise
                except Exception as e:
                    print(f"load_raw_data_ptbxl: Error during np.array conversion for 100Hz: {e}", file=sys.stderr)
                    raise

            if data is not None:
                print(f"load_raw_data_ptbxl: Attempting to save {raw_npy_path}...") # DEBUG
                try:
                    with open(raw_npy_path, 'wb') as f_out:
                        pickle.dump(data, f_out, protocol=4)
                    print(f"load_raw_data_ptbxl: Successfully saved {raw_npy_path}") # DEBUG
                except MemoryError:
                    print(f"load_raw_data_ptbxl: MEMORY ERROR during pickling for 100Hz. File: {raw_npy_path}", file=sys.stderr)
                    raise
                except Exception as e:
                    print(f"load_raw_data_ptbxl: Error during pickling for 100Hz: {e}", file=sys.stderr)
                    raise
            else:
                 print(f"load_raw_data_ptbxl: Data is None, skipping pickling for {raw_npy_path}.")


    elif sampling_rate == 500:
        npy_filename = f'raw500_subset{subset_size}.npy' if subset_size else 'raw500.npy'
        raw_npy_path = os.path.join(path, npy_filename)

        if os.path.exists(raw_npy_path) and subset_size is None:
            print(f"load_raw_data_ptbxl: Loading existing {raw_npy_path}") # DEBUG
            data = np.load(raw_npy_path, allow_pickle=True)
        else:
            print(f"load_raw_data_ptbxl: Reading signals for {raw_npy_path} (subset_size: {subset_size})...") # DEBUG
            signals_list = []
            for i, f_path_suffix in enumerate(tqdm(df_to_process.filename_hr, desc=f"Loading 500Hz signals (subset {subset_size})" if subset_size else "Loading 500Hz signals")):
                full_file_path = os.path.join(path, f_path_suffix)
                try:
                    signal_data, meta_data = wfdb.rdsamp(full_file_path)
                    signals_list.append(signal_data)
                except Exception as e:
                    print(f"Error reading file {i} ({full_file_path}): {e}", file=sys.stderr)
                    signals_list.append(None)

            print(f"load_raw_data_ptbxl: Finished reading {len(signals_list)} signals. Filtering None entries...") # DEBUG
            valid_signals = [s for s in signals_list if s is not None]
            del signals_list # Free memory
            gc.collect() # Force garbage collection

            if not valid_signals:
                print("WARNING: No valid signals were loaded for 500Hz. `data` will be an empty array.", file=sys.stderr)
                data = np.array([], dtype=object)
            else:
                print(f"load_raw_data_ptbxl: {len(valid_signals)} valid signals. Attempting conversion to numpy array...") # DEBUG
                try:
                    data = np.array(valid_signals, dtype=object)
                    print(f"load_raw_data_ptbxl: Conversion to numpy array complete. Shape: {data.shape if hasattr(data, 'shape') else 'N/A'}. Pickling...") # DEBUG
                except MemoryError:
                    print("load_raw_data_ptbxl: MEMORY ERROR during np.array conversion for 500Hz. Try reducing subset_size or using a high-RAM runtime.", file=sys.stderr)
                    raise
                except Exception as e:
                    print(f"load_raw_data_ptbxl: Error during np.array conversion for 500Hz: {e}", file=sys.stderr)
                    raise

            if data is not None:
                print(f"load_raw_data_ptbxl: Attempting to save {raw_npy_path}...") # DEBUG
                try:
                    with open(raw_npy_path, 'wb') as f_out:
                        pickle.dump(data, f_out, protocol=4)
                    print(f"load_raw_data_ptbxl: Successfully saved {raw_npy_path}") # DEBUG
                except MemoryError:
                    print(f"load_raw_data_ptbxl: MEMORY ERROR during pickling for 500Hz. File: {raw_npy_path}", file=sys.stderr)
                    raise
                except Exception as e:
                    print(f"load_raw_data_ptbxl: Error during pickling for 500Hz: {e}", file=sys.stderr)
                    raise
            else:
                print(f"load_raw_data_ptbxl: Data is None, skipping pickling for {raw_npy_path}.")
    else:
        raise ValueError(f"Unsupported sampling_rate: {sampling_rate} for PTB-XL")

    if data is None:
        print(f"CRITICAL WARNING: `data` is None at the end of load_raw_data_ptbxl for sampling_rate {sampling_rate}. This indicates a flaw in the loading logic.", file=sys.stderr)

    return data

def compute_label_aggregations(df, folder, ctype):
    df['scp_codes_len'] = df.scp_codes.apply(lambda x: len(x))
    aggregation_df = pd.read_csv(os.path.join(folder, 'scp_statements.csv'), index_col=0)

    if ctype in ['diagnostic', 'subdiagnostic', 'superdiagnostic']:
        diag_agg_df = aggregation_df[aggregation_df.diagnostic == 1.0]
        def aggregate_all_diagnostic(y_dic):
            tmp = []
            for key in y_dic.keys():
                if key in diag_agg_df.index:
                    tmp.append(key)
            return list(set(tmp))

        def aggregate_subdiagnostic(y_dic):
            tmp = []
            for key in y_dic.keys():
                if key in diag_agg_df.index:
                    c = diag_agg_df.loc[key].diagnostic_subclass
                    if str(c) != 'nan':
                        tmp.append(c)
            return list(set(tmp))

        def aggregate_diagnostic(y_dic):
            tmp = []
            for key in y_dic.keys():
                if key in diag_agg_df.index:
                    c = diag_agg_df.loc[key].diagnostic_class
                    if str(c) != 'nan':
                        tmp.append(c)
            return list(set(tmp))

        if ctype == 'diagnostic':
            df['diagnostic'] = df.scp_codes.apply(aggregate_all_diagnostic)
            df['diagnostic_len'] = df.diagnostic.apply(lambda x: len(x))
        elif ctype == 'subdiagnostic':
            df['subdiagnostic'] = df.scp_codes.apply(aggregate_subdiagnostic)
            df['subdiagnostic_len'] = df.subdiagnostic.apply(lambda x: len(x))
        elif ctype == 'superdiagnostic':
            df['superdiagnostic'] = df.scp_codes.apply(aggregate_diagnostic)
            df['superdiagnostic_len'] = df.superdiagnostic.apply(lambda x: len(x))

    elif ctype == 'form':
        form_agg_df = aggregation_df[aggregation_df.form == 1.0]
        def aggregate_form(y_dic):
            tmp = []
            for key in y_dic.keys():
                if key in form_agg_df.index:
                    tmp.append(key)
            return list(set(tmp))
        df['form'] = df.scp_codes.apply(aggregate_form)
        df['form_len'] = df.form.apply(lambda x: len(x))

    elif ctype == 'rhythm':
        rhythm_agg_df = aggregation_df[aggregation_df.rhythm == 1.0]
        def aggregate_rhythm(y_dic):
            tmp = []
            for key in y_dic.keys():
                if key in rhythm_agg_df.index:
                    tmp.append(key)
            return list(set(tmp))
        df['rhythm'] = df.scp_codes.apply(aggregate_rhythm)
        df['rhythm_len'] = df.rhythm.apply(lambda x: len(x))

    elif ctype == 'all':
        df['all_scp'] = df.scp_codes.apply(lambda x: list(set(x.keys())))
    return df

def select_data(XX, YY, ctype, min_samples, outputfolder):
    mlb = MultiLabelBinarizer()
    X = None
    Y = None
    y = None

    # Ensure YY (labels) corresponds to XX (data), especially if XX is a subset
    if len(XX) != len(YY):
        print(f"select_data: WARNING - Mismatch in lengths of XX ({len(XX)}) and YY ({len(YY)}).")
        if len(XX) < len(YY):
            print(f"select_data: Assuming XX is a subset, trimming YY to match first {len(XX)} records for mask creation.")
            YY_for_masking = YY.head(len(XX)).copy()
        else:
            print(f"select_data: Length of XX is not less than YY. Using YY as is for masking. This might lead to errors if XX is actually a subset from a different source.")
            YY_for_masking = YY.copy()
    else:
        YY_for_masking = YY.copy()


    if ctype == 'diagnostic':
        boolean_mask = YY_for_masking.diagnostic_len > 0
        X = XX[boolean_mask]
        Y = YY_for_masking[boolean_mask].copy()
        if not Y.empty:
            mlb.fit(Y.diagnostic.values)
            y = mlb.transform(Y.diagnostic.values)
    elif ctype == 'subdiagnostic':
        counts = pd.Series(np.concatenate(YY_for_masking.subdiagnostic.values)).value_counts()
        counts = counts[counts > min_samples]

        YY_for_masking['subdiagnostic_filtered'] = YY_for_masking.subdiagnostic.apply(lambda x: list(set(x).intersection(set(counts.index.values))))
        YY_for_masking['subdiagnostic_len_filtered'] = YY_for_masking.subdiagnostic_filtered.apply(lambda x: len(x))

        boolean_mask = YY_for_masking.subdiagnostic_len_filtered > 0
        X = XX[boolean_mask]
        Y = YY_for_masking[boolean_mask].copy()
        if not Y.empty:
            mlb.fit(Y.subdiagnostic_filtered.values)
            y = mlb.transform(Y.subdiagnostic_filtered.values)
    elif ctype == 'superdiagnostic':
        counts = pd.Series(np.concatenate(YY_for_masking.superdiagnostic.values)).value_counts()
        counts = counts[counts > min_samples]

        YY_for_masking['superdiagnostic_filtered'] = YY_for_masking.superdiagnostic.apply(lambda x: list(set(x).intersection(set(counts.index.values))))
        YY_for_masking['superdiagnostic_len_filtered'] = YY_for_masking.superdiagnostic_filtered.apply(lambda x: len(x))

        boolean_mask = YY_for_masking.superdiagnostic_len_filtered > 0
        X = XX[boolean_mask]
        Y = YY_for_masking[boolean_mask].copy()
        if not Y.empty:
            mlb.fit(Y.superdiagnostic_filtered.values)
            y = mlb.transform(Y.superdiagnostic_filtered.values)
    elif ctype == 'form':
        counts = pd.Series(np.concatenate(YY_for_masking.form.values)).value_counts()
        counts = counts[counts > min_samples]

        YY_for_masking['form_filtered'] = YY_for_masking.form.apply(lambda x: list(set(x).intersection(set(counts.index.values))))
        YY_for_masking['form_len_filtered'] = YY_for_masking.form_filtered.apply(lambda x: len(x))

        boolean_mask = YY_for_masking.form_len_filtered > 0
        X = XX[boolean_mask]
        Y = YY_for_masking[boolean_mask].copy()
        if not Y.empty:
            mlb.fit(Y.form_filtered.values)
            y = mlb.transform(Y.form_filtered.values)
    elif ctype == 'rhythm':
        counts = pd.Series(np.concatenate(YY_for_masking.rhythm.values)).value_counts()
        counts = counts[counts > min_samples]

        YY_for_masking['rhythm_filtered'] = YY_for_masking.rhythm.apply(lambda x: list(set(x).intersection(set(counts.index.values))))
        YY_for_masking['rhythm_len_filtered'] = YY_for_masking.rhythm_filtered.apply(lambda x: len(x))

        boolean_mask = YY_for_masking.rhythm_len_filtered > 0
        X = XX[boolean_mask]
        Y = YY_for_masking[boolean_mask].copy()
        if not Y.empty:
            mlb.fit(Y.rhythm_filtered.values)
            y = mlb.transform(Y.rhythm_filtered.values)
    elif ctype == 'all':
        if 'all_scp_len' not in YY_for_masking.columns:
             if 'all_scp' not in YY_for_masking.columns:
                  YY_for_masking['all_scp'] = YY_for_masking.scp_codes.apply(lambda x: list(set(x.keys())))
             YY_for_masking['all_scp_len'] = YY_for_masking.all_scp.apply(lambda x: len(x))


        counts = pd.Series(np.concatenate(YY_for_masking.all_scp.values)).value_counts()
        counts = counts[counts > min_samples]

        YY_for_masking['all_scp_filtered'] = YY_for_masking.all_scp.apply(lambda x: list(set(x).intersection(set(counts.index.values))))
        YY_for_masking['all_scp_len_filtered'] = YY_for_masking.all_scp_filtered.apply(lambda x: len(x))

        boolean_mask = YY_for_masking.all_scp_len_filtered > 0
        X = XX[boolean_mask]
        Y = YY_for_masking[boolean_mask].copy()
        if not Y.empty:
            mlb.fit(Y.all_scp_filtered.values)
            y = mlb.transform(Y.all_scp_filtered.values)
    else:
        raise ValueError(f"Unsupported ctype: {ctype} in select_data")

    if X is None or y is None or Y is None or Y.empty:
        print(f"WARNING: X, Y, or y is None or Y is empty after select_data for ctype '{ctype}'. This might mean all data was filtered out.")
        if X is None or y is None or Y is None:
             raise ValueError(f"X or y or Y is None after select_data for ctype '{ctype}'. This might mean no data remained after filtering.")


    with open(os.path.join(outputfolder, 'mlb.pkl'), 'wb') as tokenizer:
        pickle.dump(mlb, tokenizer)

    return X, Y, y, mlb

def preprocess_signals(X_train, X_validation, X_test, outputfolder, seed=None):
    ss = StandardScaler()

    valid_X_train_for_fit = [x.flatten()[:, np.newaxis] for x in X_train if hasattr(x, 'size') and x.size > 0]
    if not valid_X_train_for_fit:
        raise ValueError("X_train is empty or contains only empty/invalid signals after filtering, cannot fit StandardScaler.")

    stacked_X_train = np.vstack(valid_X_train_for_fit).astype(float)
    if stacked_X_train.size == 0:
        raise ValueError("X_train results in an empty stacked array, cannot fit StandardScaler.")
    ss.fit(stacked_X_train)

    filename = f'/standard_scaler_seed{seed}.pkl' if seed is not None else '/standard_scaler.pkl'
    with open(os.path.join(outputfolder, filename.lstrip('/')), 'wb') as ss_file:
        pickle.dump(ss, ss_file)

    X_train = apply_standardizer(X_train, ss)
    X_valid = apply_standardizer(X_validation, ss)
    X_test = apply_standardizer(X_test, ss)
    return X_train, X_valid, X_test

def apply_standardizer(X_list, ss): # Renamed X to X_list for clarity
    X_tmp = []
    # Determine the number of channels from the first valid signal if possible
    # This is used for reshaping empty arrays correctly.
    num_channels = 0
    for x_item in X_list:
        if hasattr(x_item, 'shape') and x_item.ndim == 2:
            num_channels = x_item.shape[1]
            break

    for x in X_list: # Iterate over the list of signals
        if not hasattr(x, 'size') or x.size == 0:
            # Create an empty array with 0 time steps but the correct number of channels
            X_tmp.append(np.array([]).reshape(0, num_channels if num_channels > 0 else 0))
            continue

        original_shape = x.shape # e.g., (5000, 12)

        # The scaler was fit on data of shape (n_total_points, 1)
        # So, we flatten each signal, transform, and then reshape back
        flattened_signal = x.flatten()[:, np.newaxis] # Shape becomes (5000*12, 1)
        transformed_flat_signal = ss.transform(flattened_signal) # Output shape (5000*12, 1)

        X_tmp.append(transformed_flat_signal.reshape(original_shape)) # Reshape back to (5000, 12)

    # Attempt to create a uniform numpy array;
    # For PTB-XL, signals should have consistent length after rdsamp (e.g., 5000 for 500Hz)
    # so a direct np.array should work. If they were truly jagged, dtype=object would be needed.
    try:
        return np.array(X_tmp)
    except ValueError as e:
        print(f"apply_standardizer: ValueError during np.array(X_tmp). Shapes might be inconsistent. Error: {e}", file=sys.stderr)
        print(f"apply_standardizer: Shapes in X_tmp: {[arr.shape for arr in X_tmp if hasattr(arr, 'shape')]}", file=sys.stderr)
        return np.array(X_tmp, dtype=object) # Fallback for truly jagged arrays

"""

file_path = "utils.py" # Assumes you are in /content/dnn_ecg_comparison/preparation/

try:
    with open(file_path, "w") as f:
        f.write(utils_content)
    print(f"Successfully wrote content to {file_path}")
except Exception as e:
    print(f"Error writing to {file_path}: {e}")

# Now, run the ptbxl.py script
print("Attempting to run ptbxl.py for split 1 with refined utils.py...")
!poetry run python ptbxl.py 1
```

```python

import os
from pathlib import Path
import shutil
import subprocess

print(f"Attempting to set up environment and verify core script files...")
project_root_path = Path("/content/dnn_ecg_comparison/")
experiment_dir_path = project_root_path / "experiment"
codes_dir_path = experiment_dir_path / "codes"
models_py_target_path = codes_dir_path / "models.py"
dataset_parent_dir = codes_dir_path / "data"
dataset_py_target_path = dataset_parent_dir / "dataset.py"
architectures_parent_dir = codes_dir_path / "architectures"
resnet1d_py_target_path = architectures_parent_dir / "resnet1d.py"


if not experiment_dir_path.exists():
    print(f"ERROR: Project directory {project_root_path} or {experiment_dir_path} not found. Please ensure the initial git clone was successful.")
else:
    if os.getcwd() != str(experiment_dir_path):
        os.chdir(str(experiment_dir_path))
print(f"Current directory: {os.getcwd()}")

codes_dir_path.mkdir(parents=True, exist_ok=True)
dataset_parent_dir.mkdir(parents=True, exist_ok=True)
architectures_parent_dir.mkdir(parents=True, exist_ok=True)


models_py_content = r"""
import sys
import os
import torch.nn as nn

# Add project root to sys.path to allow importing 'config'
# and to allow 'codes.architectures' to be found.
# This assumes models.py is in experiment/codes/
_current_dir = os.path.dirname(os.path.abspath(__file__))
_codes_dir = _current_dir # /experiment/codes/
_experiment_dir = os.path.join(_codes_dir, "..") # /experiment/
_project_root = os.path.join(_experiment_dir, "..") # /dnn_ecg_comparison/

if _project_root not in sys.path:
    sys.path.insert(0, _project_root)
if _experiment_dir not in sys.path: # Needed for 'from codes.architectures...'
    sys.path.insert(0, _experiment_dir)


import config # Now config should be importable
# The following imports are relative to the 'codes' package
from .architectures import resnet1d, senet1d, bi_lstm, transformer, lambdanet1d, mobilenetv3_1d, nfnet1d, effnet1d

def load_model(modelname, num_classes):
    '''Load model.

    Args:
        modelname (str): Model name to be loaded.
        num_classes (int): Number of classes for the final output layer.
    Returns:
        model (torch.nn.Module): Loaded model.
    '''
    if modelname not in config.modelname_routing.keys():
        raise Exception(f"Model name {modelname} is not defined.")

    module_name_str, function_name = config.modelname_routing[modelname]

    # Dynamically get the architectures module (e.g., codes.architectures.resnet1d)
    # __name__ will be 'codes.models' when imported
    # __name__.rsplit('.',1)[0] will be 'codes'
    parent_package_name = __name__.rsplit('.', 1)[0]
    architectures_package = getattr(sys.modules[parent_package_name], "architectures")

    model_module = getattr(architectures_package, module_name_str)
    model_constructor = getattr(model_module, function_name)

    # Instantiate the backbone model, it expects 'backbone_out_dim' or uses its default
    # It does NOT take 'num_classes' directly for the backbone part.
    print(f"DEBUG models.py: Instantiating backbone {modelname} using constructor {function_name} from {module_name_str}")
    backbone = model_constructor(backbone_out_dim=config.backbone_out_dim)

    # Create the full model by adding a final linear layer for classification
    model = nn.Sequential(
        backbone,
        nn.Linear(config.backbone_out_dim, num_classes)
    )
    print(f"DEBUG models.py: Created sequential model with backbone and final linear layer ({config.backbone_out_dim} -> {num_classes})")
    return model

if __name__ == "__main__":
    # This __main__ block is for testing models.py directly
    # It needs careful path setup if run standalone.
    print("Testing model loading directly from models.py...")
    # Add project root to path for config and architectures if run directly
    _current_dir_main = os.path.dirname(os.path.abspath(__file__))
    _codes_dir_main = _current_dir_main
    _experiment_dir_main = os.path.join(_codes_dir_main, "..")
    _project_root_main = os.path.join(_experiment_dir_main, "..")

    if _project_root_main not in sys.path:
        sys.path.insert(0, _project_root_main)
    if _experiment_dir_main not in sys.path:
         sys.path.insert(0, _experiment_dir_main) # to find 'codes'

    import config # Re-import with updated path

    try:
        modelname_test = "resnet1d-18"
        num_classes_test = 10
        print(f"Attempting to load '{modelname_test}' with {num_classes_test} classes...")
        model_instance = load_model(modelname_test, num_classes_test)
        print(f"Successfully loaded {modelname_test}: {model_instance}")
    except Exception as e:
        print(f"Error during model loading test in models.py: {e}")
        print("Current sys.path for models.py direct execution:", sys.path)

"""
print(f"\n--- Writing content to: {models_py_target_path} ---")
try:
    with open(models_py_target_path, "w") as f:
        f.write(models_py_content)
    print(f"Successfully wrote content to {models_py_target_path}")
    if models_py_target_path.stat().st_size == 0:
        print(f"WARNING: {models_py_target_path} is EMPTY after writing.")
except Exception as e:
    print(f"ERROR writing to {models_py_target_path}: {e}")

dataset_py_content = r"""
import os
import pickle
from typing import Optional, List, Tuple

import numpy as np
from torch.utils.data import Dataset

class PTBXLDataset(Dataset):

    def __init__(self, X, y, transform:Optional[List]=None):
        '''
        Args:
            X (np.ndarray): Data.
            y (np.ndarray): Labels.
            transform (List): List of transformations to be applied.
        '''
        self.data = X
        self.label = y
        self.transform = transform

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        data, label = self.data[index], self.label[index]
        if self.transform:
            sample = {"data": data, "label": label}
            sample = self.transform(sample)
            data, label = sample["data"], sample["label"]
        return data, label

class PTBXLLabelDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
    def __len__(self):
        return len(self.X)
    def __getitem__(self, idx):
        return self.X[idx], self.y[idx]
"""
print(f"\n--- Writing content to: {dataset_py_target_path} ---")
try:
    dataset_parent_dir.mkdir(parents=True, exist_ok=True)
    with open(dataset_py_target_path, "w") as f:
        f.write(dataset_py_content)
    print(f"Successfully wrote content to {dataset_py_target_path}")
    if dataset_py_target_path.stat().st_size == 0:
        print(f"WARNING: {dataset_py_target_path} is EMPTY after writing.")
except Exception as e:
    print(f"ERROR writing to {dataset_py_target_path}: {e}")

print(f"\n--- Checking for file: {resnet1d_py_target_path} ---")
if not resnet1d_py_target_path.exists() or resnet1d_py_target_path.stat().st_size == 0:
    if resnet1d_py_target_path.exists():
        print(f"File {resnet1d_py_target_path} exists but is empty/invalid. Attempting to re-download...")
        os.remove(resnet1d_py_target_path)
    else:
        print(f"CRITICAL: The file {resnet1d_py_target_path} was NOT FOUND.")

    print(f"Attempting to download resnet1d.py from GitHub using curl...")
    resnet1d_url_raw = "https://raw.githubusercontent.com/seitalab/dnn_ecg_comparison/master/experiment/codes/architectures/resnet1d.py"
    curl_command_resnet1d = f"curl -SLf -o {str(resnet1d_py_target_path)} {resnet1d_url_raw}"
    print(f"Executing: {curl_command_resnet1d}")
    process_resnet1d = subprocess.run(curl_command_resnet1d, shell=True, capture_output=True, text=True)
    if process_resnet1d.returncode == 0 and resnet1d_py_target_path.exists() and resnet1d_py_target_path.stat().st_size > 0:
        print(f"Successfully downloaded {resnet1d_py_target_path} from GitHub.")
    else:
        print(f"ERROR: curl failed for resnet1d.py (code {process_resnet1d.returncode}) or file still missing/empty.")
        print(f"  curl stdout: {process_resnet1d.stdout.strip()}")
        print(f"  curl stderr: {process_resnet1d.stderr.strip()}")
else:
    print(f"File {resnet1d_py_target_path} found and is not empty.")

if resnet1d_py_target_path.exists():
    print(f"\n--- Content of {resnet1d_py_target_path} (First 50 lines) ---")
    with open(resnet1d_py_target_path, 'r') as f_resnet1d:
        for i, line in enumerate(f_resnet1d):
            if i < 50:
                print(line, end='')
            else:
                break
    print(f"\n--- End of {resnet1d_py_target_path} content ---")
else:
    print(f"{resnet1d_py_target_path} does not exist, cannot show content.")


print("\n--- Ensuring __init__.py files exist ---")
Path(experiment_dir_path / "__init__.py").touch(exist_ok=True)
print(f"Touched {experiment_dir_path / '__init__.py'}")
Path(codes_dir_path / "__init__.py").touch(exist_ok=True)
print(f"Touched {codes_dir_path / '__init__.py'}")
Path(dataset_parent_dir / "__init__.py").touch(exist_ok=True)
print(f"Touched {dataset_parent_dir / '__init__.py'}")
Path(architectures_parent_dir / "__init__.py").touch(exist_ok=True)
print(f"Touched {architectures_parent_dir / '__init__.py'}")
print("--- __init__.py setup complete ---")


execute_script_content_final = r"""
import os
import sys
import pickle
import warnings

# --- Robust path setup for Colab execution ---
print(f"DEBUG: Initial sys.path inside execute_clf_multilabel.py: {sys.path}")
print(f"DEBUG: Initial os.getcwd() inside execute_clf_multilabel.py: {os.getcwd()}")

project_root_abs = "/content/dnn_ecg_comparison"
script_dir_abs = os.path.dirname(os.path.abspath(__file__)) # Should be /content/dnn_ecg_comparison/experiment

if project_root_abs not in sys.path:
    sys.path.insert(0, project_root_abs)
if script_dir_abs not in sys.path:
    sys.path.insert(0, script_dir_abs)

print(f"DEBUG: Modified sys.path inside execute_clf_multilabel.py: {sys.path}")
# --- End of robust path setup ---

import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import roc_auc_score, f1_score
from tqdm import tqdm

import config
from preparation import utils
from codes.models import load_model
from codes.data.dataset import PTBXLDataset, PTBXLLabelDataset

warnings.simplefilter("ignore")

class MultiLabelClassifier:
    def __init__(self, modelname, task, device, split_number,
                 data_dir, result_dir, batch_size=None, learning_rate=None):
        self.modelname = modelname
        self.task = task
        self.device = device
        self.split_number = split_number
        self.data_dir = data_dir
        self.result_dir = result_dir
        self.num_epochs = config.num_epochs
        self.eval_every = config.eval_every
        self.patience = config.patience
        self.criterion = nn.BCEWithLogitsLoss()

        self.data_split_string = utils.form_datasplit_string(split_number)
        self.output_dir = os.path.join(self.result_dir, self.task, self.data_split_string, self.modelname)
        os.makedirs(self.output_dir, exist_ok=True)
        print(f"MultiLabelClassifier: Initialized. Results will be saved to: {self.output_dir}")

        self.batch_size = batch_size
        self.learning_rate = learning_rate
        if self.batch_size is None or self.learning_rate is None:
            print("MultiLabelClassifier: Batch size or learning rate not provided, attempting to load from grid search...")
            loaded_bs, loaded_lr = self._select_hyperparams()
            if self.batch_size is None:
                self.batch_size = loaded_bs
            if self.learning_rate is None:
                self.learning_rate = loaded_lr
        print(f"MultiLabelClassifier: Using Batch Size = {self.batch_size}, Learning Rate = {self.learning_rate}")


    def _load_gridsearch_result(self):
        ''' Load gridsearch result from file '''
        target_file = os.path.join(config.gridsearch_result_loc, f"{self.modelname}.csv")
        print(f"MultiLabelClassifier: Attempting to load grid search results from {target_file}")
        if not os.path.exists(target_file):
             print(f"MultiLabelClassifier: Grid search result file not found: {target_file}")
             return None
        try:
            gridsearch_result = pd.read_csv(target_file)
            print(f"MultiLabelClassifier: Successfully loaded grid search results.")
            return gridsearch_result
        except Exception as e:
            print(f"MultiLabelClassifier: Error loading grid search result file {target_file}: {e}")
            return None

    def _select_hyperparams(self):
        ''' Select best hyperparameters based on grid search result '''
        if self.modelname == "resnet1d-18":
            print("MultiLabelClassifier: Using hardcoded hyperparameters for resnet1d-18 (bs=64, lr=0.001)")
            return 64, 0.001

        gridsearch_result = self._load_gridsearch_result()
        if gridsearch_result is None:
             print("MultiLabelClassifier: Grid search file missing or failed to load. Using default fallback hyperparameters (bs=128, lr=0.001).")
             return 128, 0.001

        best_params = gridsearch_result.iloc[gridsearch_result[config.selection_criteria].idxmax()]
        batch_size = int(best_params.batch_size)
        learning_rate = float(best_params.learning_rate)
        print(f"MultiLabelClassifier: Selected hyperparameters from file: bs={batch_size}, lr={learning_rate}")
        return batch_size, learning_rate

    def _load_data(self):
        ''' Load preprocessed data '''
        print("MultiLabelClassifier: Loading preprocessed data...")
        data_path = os.path.join(self.data_dir, self.task, self.data_split_string)
        print(f"MultiLabelClassifier: Data path: {data_path}")

        try:
            X_train = np.load(os.path.join(data_path, "X_train.npy"), allow_pickle=True)
            X_val = np.load(os.path.join(data_path, "X_val.npy"), allow_pickle=True)
            X_test = np.load(os.path.join(data_path, "X_test.npy"), allow_pickle=True)
            y_train = np.load(os.path.join(data_path, "y_train.npy"), allow_pickle=True)
            y_val = np.load(os.path.join(data_path, "y_val.npy"), allow_pickle=True)
            y_test = np.load(os.path.join(data_path, "y_test.npy"), allow_pickle=True)
            print("MultiLabelClassifier: Loaded data arrays successfully.")
        except FileNotFoundError as e:
             print(f"MultiLabelClassifier: ERROR - Preprocessed data file not found: {e}")
             print(f"MultiLabelClassifier: Please ensure '{config.dirname_ptbxl}/preparation/ptbxl.py {self.split_number}' was run successfully.")
             raise

        try:
            with open(os.path.join(data_path, "mlb.pkl"), "rb") as f:
                self.mlb = pickle.load(f)
            self.num_classes = len(self.mlb.classes_)
            print(f"MultiLabelClassifier: Loaded LabelBinarizer. Num classes: {self.num_classes}")
        except FileNotFoundError as e:
            print(f"MultiLabelClassifier: ERROR - mlb.pkl not found: {e}")
            print(f"MultiLabelClassifier: Please ensure '{config.dirname_ptbxl}/preparation/ptbxl.py {self.split_number}' was run successfully.")
            raise

        print("MultiLabelClassifier: Creating Datasets and DataLoaders...")
        train_dataset = PTBXLLabelDataset(X_train, y_train)
        val_dataset = PTBXLLabelDataset(X_val, y_val)
        test_dataset = PTBXLLabelDataset(X_test, y_test)

        self.train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=self.batch_size, shuffle=True)
        self.val_loader = torch.utils.data.DataLoader(val_dataset, batch_size=self.batch_size, shuffle=False)
        self.test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=self.batch_size, shuffle=False)
        print("MultiLabelClassifier: DataLoaders created.")


    def _build_model(self):
        ''' Build model, optimizer '''
        print(f"MultiLabelClassifier: Building model '{self.modelname}'...")
        self.model = load_model(self.modelname, self.num_classes)
        self.model.to(self.device)
        self.optimizer = optim.Adam(self.model.parameters(), lr=self.learning_rate)
        print("MultiLabelClassifier: Model and optimizer built.")


    def _train_epoch(self):
        ''' Train model for one epoch '''
        self.model.train()
        total_loss = 0
        for inputs, targets in self.train_loader:
            inputs = inputs.to(self.device, dtype=torch.float)
            targets = targets.to(self.device, dtype=torch.float)

            # Ensure input is (batch_size, channels, sequence_length)
            if inputs.ndim == 3 and inputs.shape[1] != 12: # Assuming 12 is num_channels
                inputs = inputs.permute(0, 2, 1)

            self.optimizer.zero_grad()
            outputs = self.model(inputs)
            loss = self.criterion(outputs, targets)
            loss.backward()
            self.optimizer.step()
            total_loss += loss.item()
        return total_loss / len(self.train_loader)

    def _evaluate(self, data_loader):
        ''' Evaluate model performance '''
        self.model.eval()
        all_targets = []
        all_outputs = []
        with torch.no_grad():
            for inputs, targets in data_loader:
                inputs = inputs.to(self.device, dtype=torch.float)
                targets = targets.to(self.device, dtype=torch.float)

                # Ensure input is (batch_size, channels, sequence_length)
                if inputs.ndim == 3 and inputs.shape[1] != 12: # Assuming 12 is num_channels
                     inputs = inputs.permute(0, 2, 1)

                outputs = self.model(inputs)
                all_targets.append(targets.cpu().numpy())
                all_outputs.append(outputs.cpu().numpy())

        all_targets = np.concatenate(all_targets, axis=0)
        all_outputs = np.concatenate(all_outputs, axis=0)
        all_probs = 1 / (1 + np.exp(-all_outputs))
        try:
             auc_scores = [roc_auc_score(all_targets[:, i], all_probs[:, i])
                           for i in range(self.num_classes) if len(np.unique(all_targets[:, i])) > 1]
             macro_auc = np.mean(auc_scores) if auc_scores else 0.0
        except ValueError as e:
             print(f"MultiLabelClassifier: Warning during AUC calculation: {e}. Setting AUC to 0.")
             macro_auc = 0.0
        return macro_auc, all_targets, all_outputs

    def run(self):
        ''' Run the training and evaluation loop '''
        self._load_data()
        self._build_model()

        best_val_auc = 0.0
        epochs_no_improve = 0
        history = {'train_loss': [], 'val_auc': []}

        print("MultiLabelClassifier: Starting training loop...")
        for epoch in range(1, self.num_epochs + 1):
            train_loss = self._train_epoch()
            history['train_loss'].append(train_loss)
            print(f"Epoch {epoch}/{self.num_epochs}, Train Loss: {train_loss:.4f}", end='')

            if epoch % self.eval_every == 0:
                val_auc, _, _ = self._evaluate(self.val_loader)
                history['val_auc'].append(val_auc)
                print(f", Val AUC: {val_auc:.4f}", end='')

                if val_auc > best_val_auc:
                    best_val_auc = val_auc
                    epochs_no_improve = 0
                    torch.save(self.model.state_dict(), os.path.join(self.output_dir, 'best_model.pth'))
                    print(" (New best validation AUC, saving model)")
                else:
                    epochs_no_improve += 1
                    print("")
                    if epochs_no_improve >= self.patience:
                        print(f"Early stopping triggered after {epoch} epochs.")
                        break
            else:
                 print("")

        print("MultiLabelClassifier: Training finished.")

        best_model_path = os.path.join(self.output_dir, 'best_model.pth')
        if os.path.exists(best_model_path):
            print("MultiLabelClassifier: Loading best model for test evaluation...")
            self.model.load_state_dict(torch.load(best_model_path))
            test_auc, test_targets, test_outputs = self._evaluate(self.test_loader)
            print(f"MultiLabelClassifier: Test AUC: {test_auc:.4f}")
        else:
            print(f"MultiLabelClassifier: No best model saved (best_model.pth not found at {best_model_path}). Skipping test evaluation.")
            test_auc = 0.0
            test_targets = np.array([])
            test_outputs = np.array([])

        print("MultiLabelClassifier: Saving results...")
        results = {'test_auc': test_auc, 'best_val_auc': best_val_auc, 'history': history}
        with open(os.path.join(self.output_dir, 'results.pkl'), 'wb') as f:
            pickle.dump(results, f)
        np.save(os.path.join(self.output_dir, 'test_targets.npy'), test_targets)
        np.save(os.path.join(self.output_dir, 'test_outputs_logits.npy'), test_outputs)
        print("MultiLabelClassifier: Results saved.")


if __name__ == "__main__":
    if len(sys.argv) < 3:
        print("Usage: python execute_clf_multilabel.py <modelname> <device> [split_number]")
        sys.exit(1)

    modelname = sys.argv[1]
    device = sys.argv[2]
    split_number = int(sys.argv[3]) if len(sys.argv) > 3 else 1

    print(f"Executing multi-label classification for model: {modelname}, device: {device}, split: {split_number}")

    task = "all"
    data_dir = os.path.join(config.data_root, config.dirname_ptbxl, "processed")
    result_dir = config.multilabel_result_loc

    mce = MultiLabelClassifier(modelname, task, device, split_number, data_dir, result_dir)
    mce.run()
    print("Script finished.")

"""

file_path_script_to_run = "execute_clf_multilabel.py"
try:
    with open(file_path_script_to_run, "w") as f:
        f.write(execute_script_content_final)
    print(f"\nSuccessfully wrote content to {file_path_script_to_run}")
except Exception as e:
    print(f"\nError writing to {file_path_script_to_run}: {e}")

print("\nAttempting to run execute_clf_multilabel.py with hardcoded hyperparameters and corrected import...")
!poetry run python execute_clf_multilabel.py resnet1d-18 cuda:0 1

```
